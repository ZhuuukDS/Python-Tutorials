# **`matplotlib`** - creating plots and figures
______________________________

```python
# import a module
from pylab import * # imports everything (easy way)
import matplotlib.pyplot as plt # importing matplotlib module under the name plt
import numpy as np # need to import numpy as well for arrays
```

```python
# This enables to show figures embedded in the notebook,
# instead of opening a new window every time
%matplotlib inline
```

### **`Examples of plots`**

```python
# coordinates of dots for plotting
x = np.linspace(0, 5, 10)
y = x ** 2
```

```python
# creating a figure from the dots
# MATLAB-like way
figure() # create a global figure instance
plot(x, y, 'r') # 'r' for type and color of the line
xlabel('x') # name of the x-axes
ylabel('y') # name of the y-axes
title('The plot') # name of the plot
show() # command to show the plot
```

```python
# another way of creating plots
# Object-oriented way
fig = plt.figure() # we store a newly created figure instance to a variable and from it...
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # ...we create a new axis instance (range from 0 to 1)

axes.plot(x, y, 'r')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('The plot');
```

```python
# two plots on one figure
fig = plt.figure()

axes1 = fig.add_axes([.1, .1, .8, .8]) # main axes
axes2 = fig.add_axes([.2, .5, .4, .3]) # insert axes

# main figure
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('The plot')

# insert
axes2.plot(y, x, 'g--')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('small plot');
```

```python
# several plots close to each other (subplots)
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('y')
    ax.set_ylabel('x')
    ax.set_title('plot')
    
fig.tight_layout()
```

```python
# defining a plot size
fig, axes = plt.subplots(figsize = (12,4)) # size in inches

axes.plot(x, y, 'r')
axes.set_xlabel('y')
axes.set_ylabel('x')
axes.set_title('plot');
```

```python
fig, ax = subplots()

ax.plot(x, x**2, label='y = x**2')
ax.plot(x, x**3, label='y = x**3')
ax.legend(loc=2)# upper left corner
ax.set_xlabel('y')
ax.set_ylabel('x')
ax.set_title('plot');
```

```python
# histograms
n = np.random.randn(100000)
fig, axes = plt.subplots(1, 2, figsize=(12,4))

axes[0].hist(n)
axes[0].set_title('Default histogram')
axes[0].set_xlim((min(n), max(n)))

axes[1].hist(n, cumulative=True, bins=50)
axes[1].set_title('Cumulative detailed histogram')
axes[1].set_xlim((min(n), max(n)));
```

### **`Drawing geometric shapes`**

Flat geometric figures can be created using `matplotlib.patches` library

```python
from matplotlib.patches import Circle, Wedge, Polygon, Ellipse, Arc, Path, PathPatch
```

1. Create a window with coordinates **xlim()**, **ylim()**
2. Create a space for a figure linked with axis **gca()**
3. Create a geometric figure (circle in this case) **Circle((a,b),c))**
4. Add the figure into a space **add_patch()**
5. Show a geometric figure **show()**

```python
plt.xlim(0, 12)
plt.ylim(0, 12)
ax = plt.gca()
circle = Circle((6, 7), 4) # ((a, b), c), where a, b - circle center coordinates, c - radius
ax.add_patch(circle)
plt.show()
```

Different shapes available

| shapes | description |
|---|---|
| __`Rectangle((x, y), width, height)`__ | a rectangle with left upper angle is at `(x, y)` having `width` and `height` |
| __`Circle((x, y), radius)`__ | a circle with its center at `(x, y)` and with radius equal to `radius` |
| __`Ellipse((x, y), width, height)`__ | an ellipse with its center at `(x, y)` and with horizontal diameter `width` and vertical diameter `height` |
| __`Wedge((x, y), radius, t1, t2)`__ | a sector with its center at `(x, y)`, with radius equal to `radius` and limited by lines of angles `t1` and `t2` |
| __`Arc((x, y), width, height, angle, t1, t2)`__ | an arc with its center at `(x, y)`, with horizontal diameter `width` and vertical diameter `height`, with the angle of rotation relative to the center `angle` and limited by lines of angles `t1` and `t2` |
| __`Polygon([(x0, y0), (x1, y1), (x2, y2), ...], closed)`__ | broken line by points `(x0, y0), (x1, y1), (x2, y2)`. If `closed=True` the 1st point connects to the last, otherwise (`closed=False`) the broken line is not closed |
| __`Path(vertices, codes)`__ | a path through points `vertices` (a list of coordinates) having path `codes` for drawing styles of paths between points (STOP (code 0), MOVETO (code 1), LINETO (code 2), CURVE3 (code 3), CURVE4 (code 4), CLOSEPOLY (code 79) |
| __`PathPatch(path, lw=3)`__ | a figure based on the `path` with line size 3 (`lw=3`) |

Drawing lines joined in the path

```python
plt.xlim(0, 8)
plt.ylim(0, 8)
ax = plt.gca()

vertices = [(0,6),(2,8),(2,4),(4,6),(4,2),(6,4),(6,0),(8,2)] # a list of points in the path
codes = [1,2,1,2,1,2,1,4] # a list of codes
path = Path(vertices, codes) # creating a path
path_patch = PathPatch(path, lw=3) # creating a figure from the path
ax.add_patch(path_patch) # adding our path into the space
ax.axes.set_axis_off() # removing axis
plt.show()
```

Drawing a cat

```python
# 1. Import module and figures from the library
from matplotlib.patches import Circle, Wedge, Polygon, Ellipse, Arc, Path, PathPatch
import matplotlib.pyplot as plt

# 2. Create a function for drawing a figure with a drawing space as a parameter
def draw_cat(ax):
 # body
    poly = [(3, 1), (4, 14), (5, 11), (8, 11), (9, 14), (10, 1)]
    polygon = Polygon(poly, fc="grey", ec="black", lw=4)
    ax.add_patch (polygon)

    # eyes
    circle = Circle((5.3, 8.5), 1.2, fc="white", ec="black", lw=4)
    ax.add_patch (circle)

    circle = Circle((7.7, 8.5), 1.2, fc="white", ec="black", lw=4)
    ax.add_patch (circle)

    # pupils
    circle = Circle((6, 8.3), 0.1, fc="black", ec="black", lw=4)
    ax.add_patch (circle)

    circle = Circle((7, 8.3), 0.1, fc="black", ec="black", lw=4)
    ax.add_patch (circle)

    # nose
    circle = Circle((6.5, 7.5), 0.3, fc="black", ec="black", lw=4)
    ax.add_patch (circle)

    # back legs
    wedge = Wedge((3, 1), 2, 86, 180, fc="grey", ec="black", lw=4)
    ax.add_patch (wedge)
    
    wedge = Wedge((10, 1), 2, 0, 94, fc="grey", ec="black", lw=4)
    ax.add_patch (wedge)

    # front legs
    ellipse = Ellipse((5.5,1.2), 2, 1.5, fc="grey", ec="black", lw=4)
    ax.add_patch (ellipse)

    ellipse = Ellipse((7.5,1.2), 2, 1.5, fc = "grey", ec="black", lw=4)
    ax.add_patch (ellipse)

    # smile
    arc =  Arc((6.5, 6.5), 5, 3, 0, 200, 340, lw=4, fill=False)
    ax.add_patch(arc)

    # line between nose and smile, mustaches
    vertices = [(6.5, 5), (6.5, 7.5), (10, 6), (6.5, 7.5), (10, 6.5), (6.5, 7.5), (10, 7),
                (6.5, 7.5), (3, 6), (6.5, 7.5), (3, 6.5), (6.5, 7.5), (3, 7)]
    
    # code 1 is for matplotlib.path.Path.MOVETO
    # число 2 is for matplotlib.path.Path.LINETO
    codes = [1, 2, 2, 1, 2, 1, 2, 1, 2, 1, 2, 1, 2]

    path = Path(vertices, codes)
    path_patch = PathPatch(path, fill=False, lw=1)
    ax.add_patch(path_patch)

# 3. Determine the size and coordinates of angles for drawing space

n = 13
m  = 15
plt.xlim(0, n)
plt.ylim(0, m)

# 4. Create a space linked with the axis, where the flat figures will be drawn (ax), call the draw function (draw_cat)

ax = plt.gca()
draw_cat(ax)

# 5. Delete the axis, coordinates and show the figure

ax.axes.set_axis_off()
plt.show()
```
